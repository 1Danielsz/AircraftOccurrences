# Drop Unnecessary Columns

```python
aircraft_drop = [
  'aeronave_matricula', 'aeronave_motor_tipo', 'aeronave_motor_quantidade', 'aeronave_tipo_icao',	'aeronave_pmd',	'aeronave_pmd_categoria',	'aeronave_assentos',	'aeronave_ano_fabricacao',	'aeronave_pais_fabricante',	
  'aeronave_pais_registro', 'aeronave_registro_categoria',	'aeronave_registro_segmento',	'aeronave_voo_origem',	'aeronave_voo_destino',	'aeronave_tipo_operacao'
]

aircraft.drop(aircraft_drop, axis = 1, inplace = True)
aircraft.head()
```

```python
occurrence_drop = [
  'codigo_ocorrencia1',	'codigo_ocorrencia2',	'codigo_ocorrencia3',	'codigo_ocorrencia4',	'investigacao_aeronave_liberada',	'investigacao_status', 'divulgacao_relatorio_numero', 
  'divulgacao_relatorio_publicado',	'divulgacao_dia_publicacao',	'total_recomendacoes',	'ocorrencia_saida_pista'
]

occurrence.drop(occurrence_drop, axis = 1, inplace= True)
occurrence.head()
```

```python
occurrence_type_drop = [
  'ocorrencia_tipo_categoria', 'taxonomia_tipo_icao'
]

occurrence_type.drop(occurrence_type_drop, axis = 1, inplace = True)
occurrence_type.head()
```

```python
contributor_factor_drop = [
  'fator_nome',	'fator_aspecto', 'fator_condicionante'	
]

contributor_factor.drop(contributor_factor_drop, axis = 1, inplace = True)
contributor_factor.head()
```

# Rename Datasets

```python
aircraft_rename = {
    'codigo_ocorrencia2': 'occurrence_id',
    'aeronave_operador_categoria': 'operator_category',
    'aeronave_tipo_veiculo': 'aircraft_type',
    'aeronave_fabricante': 'aircraft_manufacturer',
    'aeronave_modelo': 'aircraft_model',
    'aeronave_fase_operacao': 'operation_phase',
    'aeronave_nivel_dano': 'aircraft_damage_level',
    'aeronave_fatalidades_total': 'total_aircraft_fatalities'
}

aircraft.rename(columns = aircraft_rename, inplace = True)
aircraft.sample(5)
```

```python
occurrence_rename = {
    'codigo_ocorrencia': 'occurrence_id',
    'ocorrencia_classificacao': 'classification',
    'ocorrencia_latitude': 'latitude',
    'ocorrencia_longitude': 'longitude',
    'ocorrencia_cidade': 'city',
    'ocorrencia_uf': 'state',
    'ocorrencia_pais': 'country',
    'ocorrencia_aerodromo': 'aerodrome',
    'ocorrencia_dia': 'date',
    'ocorrencia_hora': 'time',
    'total_aeronaves_envolvidas': 'total_aircraft_involved'
}

occurrence.rename(columns=occurrence_rename, inplace=True)
occurrence.head()
```

```python
occurrence_type_rename = {
    'codigo_ocorrencia1': 'occurrence_id',
    'ocorrencia_tipo': 'type_category',
}

occurrence_type.rename(columns=occurrence_type_rename, inplace=True)
occurrence_type.head()
```

```python
contributor_factor_rename = {
    'codigo_ocorrencia3': 'occurrence_id',
    'fator_area': 'factor_area',
}

contributor_factor.rename(columns=contributor_factor_rename, inplace=True)
contributor_factor.head()
```
